HAI:

The housing adequacy problem only exists when there’s new demand for housing —
and that demand primarily comes from population growth (through migration, births, or demographic change).
If a province’s population doesn’t grow (Δpop = 0) or even declines (Δpop < 0), then:
there are no additional households to accommodate,
any new housing starts represent replacement or upgrades, not shortage relief,
and thus, a “housing adequacy ratio” (starts vs. required homes) isn’t meaningful.
